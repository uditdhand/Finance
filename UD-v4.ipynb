```python
fundamentals = fundamentals_temp.groupby(['GVKEY','fyear']).sum()
fundamentals['xrd'] = fundamentals['Total Annual Cash Compensation']
fundamentals['xrd'] = fundamentals['xrd'].fillna(0)
fundamentals = fundamentals.sort_values(by=['GVKEY' ,'fyear'])

fundamentals['xrd_sort'] =    fundamentals.groupby('GVKEY')['xrd'].shift(1).fillna(0)

fundamentals['total_comp'] =fundamentals['Total Annual Cash Compensation'] + fundamentals['Total Value of Options']  +   fundamentals['Total Value of Stock Awards'] + fundamentals['Present Value of Accumulated Benefits']
fundamentals['total_comp'] = fundamentals.groupby('GVKEY')['total_comp'].shift(1).fillna(0)
fundamentals = fundamentals[fundamentals['total_comp'] > 0]
fundamentals['xrd_sort'] = fundamentals['xrd_sort'] / fundamentals['total_comp']


fundamentals = fundamentals.dropna(subset=['sector'])

# normalize xrd_sort by industry: Example: xrd_sort_individual_company = xrd_sort - mean(xrd_sort_company_sector) / std(xrd_sort_company_sector)
for sector in fundamentals['sector'].unique():
    fundamentals.loc[fundamentals['sector'] == sector, 'xrd_sort'] = (fundamentals.loc[fundamentals['sector'] == sector, 'xrd_sort'] - fundamentals.loc[fundamentals['sector'] == sector, 'xrd_sort'].mean()) / fundamentals.loc[fundamentals['sector'] == sector, 'xrd_sort'].std()

fundamentals['quintile'] = fundamentals[fundamentals['xrd_sort'] != 0].groupby('fyear')['xrd_sort'].transform(lambda x: pd.qcut(x, 5, labels=False))

# keep only xrd_sort and quintile
fundamentals = fundamentals[['xrd_sort','quintile']].dropna(subset=['quintile'])
```

```python
# Convert datadate to datetime
returns['fyear'] = returns['datadate'].apply(lambda x: int(str(x)[:4]))
# Filter data for years and later
returns = returns[returns['fyear'] >= 2001]
```

```python
returns['cshoq'] = returns.groupby(['GVKEY','iid','fyear'])['cshoq'].transform('mean')
returns = returns.dropna(subset=['cshoq'])
```

```python
df = pd.merge(returns, fundamentals, on=['GVKEY', 'fyear'], how='left', suffixes=('_return', '_fundamental'))
df['datadate_return'] = df['datadate']
df = df.sort_values(by=['GVKEY', 'datadate_return'])
df.set_index(['GVKEY','datadate_return'], inplace=True)
df['mkt_cap'] = df['prccm'] * df['cshoq']
```

```python
# find top 1000 companies by market cap for each year
top_1000 = df.groupby('fyear')['mkt_cap'].nlargest(1000).reset_index()
```

```python
# remove gvkeys in top_1000 for each year from df
df_sc = df.copy()
df_sc = df_sc.reset_index()
df_sc = df_sc[~df_sc.set_index(['GVKEY','fyear']).index.isin(top_1000.set_index(['GVKEY','fyear']).index)]
df_sc = df_sc.set_index(['GVKEY','datadate_return'])
```

```python
filter =(df['exchg'] >= 11) \
        & (df['exchg'] <= 19) \
        & (df['fic'].str.contains('USA')) \
        & (df['mkt_cap'] >= 10) \
        & (df['trt1m']>= -10000)\
        & (df['trt1m']<= 10000)
        #& (~((df['sic'] >= 6000) & (df['sic'] <= 6999))) \ & ~((df['sic'] == 2834)) \

df_filtered = df[filter]
df_filtered = df_filtered.dropna(subset=['xrd_sort'])
df_filtered = df_filtered[['trt1m', 'xrd_sort', 'quintile','fyear','mkt_cap','spcseccd', 'sector']]

df_filtered_sc = df_sc[filter]
df_filtered_sc = df_filtered_sc.dropna(subset=['xrd_sort'])
df_filtered_sc = df_filtered_sc[['trt1m', 'xrd_sort', 'quintile','fyear','mkt_cap','spcseccd', 'sector']]
```

```python

df_filtered['market_cap_weight'] = df_filtered.groupby(['quintile', 'datadate_return'])['mkt_cap'].transform(
    lambda x: x / x.sum())
df_filtered['weighted_return'] = df_filtered['market_cap_weight'] * df_filtered['trt1m']

df_filtered_sc['market_cap_weight'] = df_filtered_sc.groupby(['quintile', 'datadate_return'])['mkt_cap'].transform(
    lambda x: x / x.sum())
df_filtered_sc['weighted_return'] = df_filtered_sc['market_cap_weight'] * df_filtered_sc['trt1m']
```

```python
def average_return(periods,  weight):
    result = pd.DataFrame()
    for period in periods:
        df_period = df_filtered.loc[(slice(None), slice(period[0], period[1])), :]
        df_period_sc = df_filtered_sc.loc[(slice(None), slice(period[0], period[1])), :]
        if weight == 'equal':
            result[period[0] + ' to ' + period[1]] = (df_period.groupby(['quintile', 'datadate_return'])['trt1m'].mean()).groupby('quintile').mean()
        elif weight == 'market_cap':
            result[period[0] + ' to ' + period[1]] = (df_period.groupby(['quintile', 'datadate_return'])['weighted_return'].sum()).groupby('quintile').mean()
        elif weight == 'market_cap_smallcap':
            result[period[0] + ' to ' + period[1]] = (df_period_sc.groupby(['quintile', 'datadate_return'])['weighted_return'].sum()).groupby('quintile').mean()

    
    rounded_result = result.round(2)
    #rename(columns={0: 'Quintile 1', 1: 'Quintile 2', 2: 'Quintile 3', 3: 'Quintile 4', 4: 'Quintile 5'}, inplace=True)
    ret = rounded_result.T
    ret.rename(columns={0: 'Quintile 1', 1: 'Quintile 2', 2: 'Quintile 3', 3: 'Quintile 4', 4: 'Quintile 5'}, inplace=True)
    ret['weight'] = weight
    return ret

insample = ('2003-01-01','2022-12-31')
half1 = ('2003-01-01','2012-12-31')
half2 = ('2013-01-01','2022-12-31')
periods = [half1, half2, insample]
```

```python
df_ = pd.DataFrame()
for weight in ['equal', 'market_cap', 'market_cap_smallcap']:
    df_ = pd.concat([df_, average_return(periods, weight)])

df_
```

```python
ff = pd.read_csv('F-F_Research_Data_Factors.csv')
ff['date'] = pd.to_datetime(ff['date'], format='%Y%m')
ff.set_index('date', inplace=True)
ff.index = ff.index + pd.offsets.MonthEnd(0)
ff.index.rename('datadate', inplace=True)
```

```python
def regress(returns):
    X = ff.loc[returns.index, 'Mkt-RF']
    X = sm.add_constant(X)
    y = returns
    X = X.reset_index(drop=True)
    y = y.reset_index(drop=True)
    model = sm.OLS(y, X)
    model.fit()
    # return const coef and its t-stat
    return model.fit().params[0], model.fit().tvalues[0]

def regress_3f(returns):
    X = ff.loc[returns.index, ['Mkt-RF', 'SMB', 'HML']]
    X = sm.add_constant(X)
    y = returns
    X = X.reset_index(drop=True)
    y = y.reset_index(drop=True)
    model = sm.OLS(y, X)
    model.fit()
    return model.fit().params[0], model.fit().tvalues[0]

def long_short_returns(period, df, weight, df_sc = df_filtered_sc):
    if weight == 'equal':
        df_period = df.loc[(slice(None), slice(period[0], period[1])), :]
        long_returns = df_period[df_period['quintile'] == 4]['trt1m'].groupby('datadate_return').mean()
        short_returns = df_period[df_period['quintile'] == 0]['trt1m'].groupby('datadate_return').mean()

    elif weight == 'market_cap':
        df_period = df.loc[(slice(None), slice(period[0], period[1])), :]
        long_returns = df_period[df_period['quintile'] == 4]['weighted_return'].groupby('datadate_return').sum()
        short_returns = df_period[df_period['quintile'] == 0]['weighted_return'].groupby('datadate_return').sum()

    elif weight == 'market_cap_smallcap':
        df_period = df_sc.loc[(slice(None), slice(period[0], period[1])), :]
        long_returns = df_period[df_period['quintile'] == 4]['weighted_return'].groupby('datadate_return').sum()
        short_returns = df_period[df_period['quintile'] == 0]['weighted_return'].groupby('datadate_return').sum()

    return long_returns - short_returns

def alpha(period, df, weight):
    returns = long_short_returns(period, df, weight)
    one_alpha, one_tstat = regress(returns)
    three_alpha, three_tstat = regress_3f(returns)
    result = pd.DataFrame({'1-factor-alpha': [one_alpha], '1-factor t-stat': [one_tstat], '3-factor alpha': [three_alpha], '3-factor t-stat': [three_tstat]})
    rounded_result = result.round(2)
    rounded_result['weight'] = weight
    return rounded_result

def sharpe(return_series):
    annual_shape = return_series.mean() / return_series.std() * np.sqrt(12)
    return annual_shape

df_ = pd.DataFrame()
for weight in ['equal', 'market_cap','market_cap_smallcap']:
    df_ = pd.concat([df_, alpha(insample, df_filtered, weight)])

df_
```

```python
sharpe_equal = sharpe(long_short_returns(insample, df_filtered, 'equal'))
sharpe_market_cap = sharpe(long_short_returns(insample, df_filtered, 'market_cap'))

for weight in ['equal', 'market_cap','market_cap_smallcap']:
    print('Sharpe Ratio for', weight, 'weighted returns:')
    print(round(sharpe(long_short_returns(insample, df_filtered, weight)), 2))
```
